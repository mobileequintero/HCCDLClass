```python
# Basic TensorFlow Demo in Python (Jupyter Notebook)

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image

# Load pre-trained MobileNetV2 model
model = MobileNetV2(weights='imagenet')

# Load an image (replace with your own image path, e.g., "sample.jpg")
img_path = 'BuenoCar.jpg'
img = image.load_img(img_path, target_size=(224, 224))

# Convert to array
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)

# Make prediction
preds = model.predict(x)
decoded = decode_predictions(preds, top=1)[0][0]  # Top-1 prediction

# Show results
plt.imshow(img)
plt.title(f"Prediction: {decoded[1]} ({decoded[2]*100:.2f}%)")
plt.axis("off")
plt.show()
```
